```python
coln = ['gender','signup_method', 'signup_flow', 'language', 'affiliate_channel', 'affiliate_provider', 'first_affiliate_tracked', 'signup_app','year_created','month_created','week_created', 'first_device_type', 'first_browser']
for f in coln:
    df_all_dummy = pd.get_dummies(train[f], prefix=f)
    train = pd.concat((train, df_all_dummy), axis=1)
```

```python
trainin = train.drop(coln, axis=1)
```

```python
country_destination_ = list(enumerate(np.unique(trainin.country_destination)))
country_destination_dict = { name : i for i, name in country_destination_ }
trainin.country_destination = trainin.country_destination.map(lambda x: country_destination_dict[x]).astype(int)
```

```python
aggregations = {
    'month_created': {'month_created_max': 'max'},
    'signup_flow': {'Countsignup': 'count' },
    'language': {'Countlang': 'count'},
    'affiliate_channel': {'affiliate_channel_max': 'max'},
    'affiliate_provider': {'affiliate_provider_max': 'max'},
    'first_affiliate_tracked': {'first_affiliate_tracked_max': 'max'},
    'signup_app': {'signup_app_max': 'max'},
    'first_device_type': {'first_device_type_max': 'max'},
    'first_browser': {'first_browser_max': 'max'}
}
```

```python
trainDF = trainVisitGroup.agg(aggregations)
#testDF = testVisitGroup.agg(aggregations)
```
